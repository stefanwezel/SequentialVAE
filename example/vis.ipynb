```python

colors = ['black', 'black','black','black','black']


fig, axes = plt.subplots(nrows=3,ncols=3,figsize=(8,10))


xs = np.arange(0,10, .1)


# generative process in real world
# a, b = 1, 10
# y = np.sin(a * x) + np.sin(b * x)

h01 = np.random.normal(0,1)
h11 = np.random.normal(5,1)

# y1 = np.sin(xs * h0) + np.sin(xs * h1)


h02 = np.random.normal(0,1)
h12 = np.random.normal(5,3)

y1 = np.sin(xs * h01) + np.sin(xs * h11)
y2 = np.sin(xs * h01) + np.sin(xs * h12)
y3 = np.sin(xs * h02) + np.sin(xs * h12)





axes[0][0].plot(xs, np.sin(xs * h01), linestyle=(0,(1,1)), color = colors[0])
axes[1][0].plot(xs, np.sin(xs * h01), linestyle=(0,(1,1)), color = colors[0])
axes[2][0].plot(xs, np.sin(xs * h02), linestyle=(0,(1,1)), color = colors[0])
axes[0][0].set_title(f'$h_0 = {h01:.2f}$')
axes[1][0].set_title(f'$h_0 = {h01:.2f}$')
axes[2][0].set_title(f'$h_0 = {h02:.2f}$')





axes[0][1].plot(xs, np.sin(xs * h11), linestyle=(0,(1,1)), color = colors[1])
axes[1][1].plot(xs, np.sin(xs * h12), linestyle=(0,(1,1)), color = colors[1])
axes[2][1].plot(xs, np.sin(xs * h12), linestyle=(0,(1,1)), color = colors[1])
axes[0][1].set_title(f'$h_1 = {h11:.2f}$')
axes[1][1].set_title(f'$h_1 = {h12:.2f}$')
axes[2][1].set_title(f'$h_1 = {h12:.2f}$')
#     axes[i][1].plot(xs, np.sin(xs * h1), linestyle=(0,(1,1)), color = colors[2])
# h0s = [h0]
ys = [y1,y2,y3]
# for i in range(5):
for i in range(3):




    axes[i][2].plot(xs, ys[i], linestyle=(0,(1,1)), color = colors[2])


    axes[i][0].set_ylim(-4,4)
    axes[i][1].set_ylim(-4,4)
    axes[i][2].set_ylim(-4,4)

#     axes[i][0].set_title('$sin(h_0 \cdot t)$')
#     axes[i][1].set_title('$sin(h_1 \cdot t)$')
#     axes[i][2].set_title('$x(t) = sin(h_0 \cdot t) + sin(h_1 \cdot t)$')

for i in range(3):
    for j in range(3):
        axes[i][j].set_xticks([])
        axes[i][j].set_yticks([])

# plt.savefig('../figures/example_3x3.svg')


plt.show()
```
